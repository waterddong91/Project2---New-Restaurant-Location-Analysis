```python
from yelpapi import YelpAPI
from pprint import pprint
import pandas as pd
import numpy as np

# Import API key
from config import yelp_api_key
```

## Extract

```python
yelp_api = YelpAPI(yelp_api_key)
```

```python
# test
response = yelp_api.search_query(location='toronto', limit=2, offset=1)
pprint(response['businesses'])
```

```python
neighbourhood = pd.read_csv('../Resources/Neighbourhoods.csv')
neighbourhood.head()
```

```python
# Create a set of neighbourhood lat and lng combinations
lng_lats = []

lngs = neighbourhood['LONGITUDE']
lats = neighbourhood['LATITUDE']

lng_lats = zip(lngs, lats)


offset = 1
limit = 49
total_num_queries = 5

ids = []
names = []
categories = []
ratings = []
review_counts = []
zip_code = []


for lng_lats in zip(lngs, lats):
    
    for i in range(total_num_queries):
        
        response = yelp_api.search_query(latitude=lng_lats[1], longitude=lng_lats[0], radius=5000, limit=limit, offset=offset)

        for business in range(len(response['businesses'])):
            
            try:
                ids.append(response['businesses'][business]['id'])
                names.append(response['businesses'][business]['name'])
                categories.append(response['businesses'][business]['categories'][0]['title'])
                ratings.append(response['businesses'][business]['rating'])
                review_counts.append(response['businesses'][business]['review_count'])
                zip_code.append(response['businesses'][business]['location']['zip_code'])
                
            except:
                pass
        
        offset = offset + limit
    
    offset = 1
```

```python
# Check if all the lists have the same number of elements
print(len(ids), len(names), len(ratings), len(review_counts), len(zip_code))
```

## Transform

```python
# Create a dataframe to hold the restaurant information
rating_df = pd.DataFrame({
                'id' : ids,
                'name' : names,
                'category' : categories,
                'ratings' : ratings,
                'review_counts' : review_counts,
                'zip_code' : zip_code
})
```

```python
rating_df
```

```python
# Drop the duplicated rows
yelp_rating = rating_df.drop_duplicates(subset=['id', 'name', 'ratings', 'review_counts'], keep='first')

# Rename columns
yelp_rating = yelp_rating.rename(columns={'id':'restaurant_id'})
yelp_rating = yelp_rating.set_index('restaurant_id')
```
