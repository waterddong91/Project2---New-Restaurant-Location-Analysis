```python
# Dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# %cd ../Data Cleaning
# %run crime_and_income.ipynb
income_crime_merge_left = pd.read_csv('../Data Cleaning/output_data/city_crime_income_left_140records.csv')
income_crime_merge_left.drop('Unnamed: 0', axis='columns', inplace=True)
```

```python
income_crime_merge_left.head()
```

```python
starbucks_final = pd.read_csv('../Data Cleaning/output_data/starbucks_final.csv')
starbucks_final.drop('Unnamed: 0', axis='columns', inplace=True)
```

```python
starbucks_final.head()
```

```python
income_crime_starbucks = pd.merge(income_crime_merge_left, starbucks_final, on='Hood_ID', how='left')
income_crime_starbucks.head()
```

```python
income_crime_starbucks.drop(['Neighbourhood'], axis='columns', inplace=True)
income_crime_starbucks.head()
```

```python
income_crime_starbucks['Number of Stores'].fillna(0, inplace=True)
income_crime_starbucks.isnull().sum()
```

```python
income_crime_starbucks = income_crime_starbucks.rename(columns={"Cities" : "Neighbourhood",
                                                                "Number of Stores" : "Number of Starbucks Stores"})
income_crime_starbucks
```

```python
# Import "Red_light file" and Merge together
red_light_csv = '../Data Cleaning/output_data/lights_final.csv'
red_light = pd.read_csv(red_light_csv)
red_light.head()
```

```python
crime_income_starbucks_redlight = pd.merge(income_crime_starbucks, red_light, on='Hood_ID', how='left')
```

```python
crime_income_starbucks_redlight = crime_income_starbucks_redlight.loc[:,["Neighbourhood_x",
                                                                         "Hood_ID",
                                                                         "Assault Rate",
                                                                         "Auto Theft Rate", 
                                                                         "Break&Enter Rate",
                                                                         "Homicide Rate",
                                                                         "Robberey Rate",
                                                                         "Theft Over Rate",
                                                                         "Total Average Rate",
                                                                         "Median_Income",
                                                                         "Average_Income",
                                                                         "Number of Starbucks Stores",
                                                                         "Number of Red Lights"]]
```

```python
crime_income_starbucks_redlight = crime_income_starbucks_redlight.rename(columns={"Neighbourhood_x" : "Neighbourhood"})
```

```python
crime_income_starbucks_redlight
```

```python
# import speeding light counts df
speeding_csv = '../Data Cleaning/output_data/speeding_count_df.csv'
speeding_count = pd.read_csv(speeding_csv)
speeding_count.head()
```

```python
# add housing data
# run housing_data
# %run housing_data.ipynb

housing_subset_df = pd.read_csv('../Data Cleaning/output_data/housing_data.csv')
housing_subset_df.drop('Unnamed: 0', axis='columns', inplace=True)

# merge housing data (housing_subset_df)
crime_income_starbucks_redlight_housing = pd.merge(crime_income_starbucks_redlight, housing_subset_df, left_on='Hood_ID', right_on='Neighbourhood Number', how='left')

# drop repeated columns
crime_income_starbucks_redlight_housing.drop(['Neighbourhood Number'], axis='columns', inplace=True)
```

```python
crime_income_starbucks_redlight_housing
```

```python
#Merge Crime + Income + Starbucks + Red Light + Speeding Count all together! (Final)
final_df = pd.merge(crime_income_starbucks_redlight_housing, speeding_count, on='Neighbourhood', how='left')
```

```python
#categories income ranges
bins = [0,25000,50000,75000,100000,125000,150000,175000,200000,225000,250000,500000]
income_range = ["<25K","25-50K","50-75K","75-100K","100-125K","125-150K","150-175K","175-200K","200-225K","225-250K","250K+"]
final_df["Income Group"] = pd.cut(final_df["Average_Income"], bins , labels = income_range)
```

```python
final_df
```

```python
final_df.to_csv('Final_Data.csv')
```

```python
edited_final_df = final_df.loc[:, ['Neighbourhood', 'Hood_ID', 'Assault Rate', 'Auto Theft Rate',
       'Break&Enter Rate', 'Homicide Rate', 'Robberey Rate', 'Theft Over Rate',
       'Total Average Rate', 'Median_Income', 'Average_Income',
       'Number of Starbucks Stores', 'Number of Red Lights', 'Population, 2016', 'Home Prices', 'Speeding', 'Income Group']]
edited_final_df.head()
```

```python
# Change NaN to 0
edited_final_df['Number of Red Lights'].fillna(0, inplace=True)
edited_final_df['Speeding'].fillna(0, inplace=True)
edited_final_df.isnull().sum()
```

```python
# change the datatype
edited_final_df['Number of Starbucks Stores'] = edited_final_df['Number of Starbucks Stores'].astype(int)
edited_final_df['Number of Red Lights'] = edited_final_df['Number of Red Lights'].astype(int)
edited_final_df['Speeding'] = edited_final_df['Speeding'].astype(int)
edited_final_df.dtypes
```
